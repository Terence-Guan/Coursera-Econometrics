```python
# import packages for the analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
%matplotlib inline

import statsmodels.formula.api as smf
```

```python
SP500 = pd.read_table('TrainExer 3-1.txt')
SP500
```

## (a) Log Transformation Model

```python
SP500['logIndex'] = np.log(SP500['Index'])
Change_logIndex = [0]
for i in range(86):
    Change_logIndex.append(SP500['logIndex'][i+1]-SP500['logIndex'][i])
SP500['Change_logIndex'] = Change_logIndex
SP500
```

```python
# create a fitted model
lm = smf.ols(formula='Change_logIndex ~ BookMarket', data=SP500).fit()

lm.summary()
```

## (b) No Transformation Model

```python
olm = smf.ols(formula='Index ~ BookMarket', data=SP500).fit()

olm.summary()
```

## (c) Residuals Scatter Plot

```python
# Residuals scatter plot for the log transformation model
## Residuals against fitted values
logfig = plt.figure(figsize=(6,6))
plt.scatter(lm.fittedvalues, lm.resid)
plt.axhline(y=0, linewidth=2, color='r')

## Residuals against predictors
logfig = plt.figure(figsize=(6,6))
plt.scatter(SP500['Change_logIndex'], lm.resid)
plt.axhline(y=0, linewidth=2, color='r')
```

```python
# Residuals scatter plot for the NO transformation model
## Residuals against fitted values
logfig = plt.figure(figsize=(6,6))
plt.scatter(olm.fittedvalues, olm.resid)
plt.axhline(y=0, linewidth=2, color='r')

## Residuals against predictors
logfig = plt.figure(figsize=(6,6))
plt.scatter(SP500['Index'], olm.resid)
plt.axhline(y=0, linewidth=2, color='r')
```

```python
# Residuals against time
logfig = plt.figure(figsize=(6,6))
plt.scatter(SP500['Year'], lm.resid)
plt.axhline(y=0, linewidth=2, color='r')

logfig = plt.figure(figsize=(6,6))
plt.scatter(SP500['Year'], olm.resid)
plt.axhline(y=0, linewidth=2, color='r')
```

A pattern and strong persistence of the residuals can be seen from the none transformation model, which means it's not $normal$.

```python
IndexRate = [0]
for i in range(86):
    IndexRate.append((SP500['Index'][i+1]-SP500['Index'][i])/SP500['Index'][i])
SP500['IndexRate'] = IndexRate
SP500
```

```python
rlm = smf.ols(formula='IndexRate ~ BookMarket', data=SP500).fit()

rlm.summary()
```

## (b) Quadratic Relation between Index and BookMarket

```python
SP500['BM_squared'] = SP500['BookMarket']**2
qlm = smf.ols(formula='Change_logIndex ~ BookMarket + BM_squared', data=SP500).fit()

qlm.summary()
```

As **P value is big**, this model is **not** quadratic

## (c) Dummy Variable of Year

```python
SP500['Y_dummy_inter'] = [0 if SP500['Year'][i]<1980 else SP500['BookMarket'][i] for i in range(len(SP500['Year']))]

ylm = smf.ols(formula='Change_logIndex ~ BookMarket + Y_dummy_inter', data=SP500).fit()

ylm.summary()
```

There is no significant relationship
